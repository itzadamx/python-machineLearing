```python
# Activity 2.1 : Run each step in IDLE and display the results
import urllib.request as req   # the request object
# Optional: run dir(req) to see the properties and methods of urllib.request module
resp=req.urlopen("https://www.python.org")   # response object
text=resp.read()                             # read the web page 
# Optional: text1=resp.read(100)              # read the 100 elements of the web page
print(text)
```

```python
# Activity 2.2
# tokenize by using python string's split method:

tokens_by_split=text.split()

import pprint
#pprint.pprint(tokens_by_split)      # this outputs everything, which could be quite big

pprint.pprint(tokens_by_split[:100]) # this outputs only up to the first 100 items 
```

```python
# Activity 3
import re

tokens_by_re=re.split(r'\W+', text)  # split on regular expression r'\W+' (r stands for 'raw')
#text=str(text,'utf8')
print("len(tokens_by_re)")  #
print(len(tokens_by_re))  #
print("Up to 100 tokens from tokens_from_re")
print(tokens_by_re[:100]) # print the first 100 tokens (if there are >100), or all tokens (if <=100)

import pprint
pprint.pprint(tokens_by_re[:100])  # pretty-printing up to 100 items
```

```python
#Activity 4
import nltk
import bs4
import pprint

clean_html_text=bs4.BeautifulSoup(text).getText()

tokens_by_cleaned_split=[token for token in clean_html_text.split()]

pprint.pprint(tokens_by_cleaned_split)
```

```python
#Activity 5.1
freq_by_nltk=nltk.FreqDist(tokens_by_cleaned_split)
freq_by_nltk
```

```python
#Activity 5.2
import operator   # we need to import this for the following to work
sorted_freq_by_nltk=sorted(freq_by_nltk.items(), key=operator.itemgetter(1), reverse=True)
pprint.pprint(sorted_freq_by_nltk)
```

```python
#Activity 5.3
freq_by_nltk.plot(30)  # this plots the first 30 frequencies
freq_by_nltk.plot(30,cumulative=True)
```
